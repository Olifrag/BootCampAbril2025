Bucle (WHILE)
While nos deja ejecutar una serie de instrucciones mientras una condicion se cumpla
Al momento que la condicion deje de ser cierta, el bucle para 

```python
numero_a_adivinar = 10
numero = 3
while numero != numero_a_adivinar:
    numero = int(input("Introduce un numero "))
    print(numero)
print ("Numero correcto")
```

```python
#El mata Pcs
numero = 10
while True:
    numero_a_adivinar = int(input("Introduce un numero"))
    if numero_a_adivinar == numero:
        print ("El numero es correcto")
        break
else:
    print("El numero es correcto")
```

```python
password=int(input("Introduce la contrasea"))
password_adivinar =123456

while password != password_adivinar:

    if password == password_adivinar:
         print("Correcto")
         break
    else:
          print("Incorrecto")
         
```

```python
lista_frutas = ["pera", "manzanas", "banana"]
#diferente tipos de datos
lista_bool=[True, False, True, True, False]
# Indices de 0 para adelante, Posicion del 1 para adelante, Indices y poscion es todo lo que esta dentro de los corchetes 
# Indice = [] Posicion = .pop
#Anadir elementos a la lista 
lista_frutas.append("Kiwi")
print(lista_frutas)
#modificar 
lista_frutas[1] = "Durazno"
print(lista_frutas)
#eliminar 
lista_frutas.remove("pera")
print(lista_frutas)
# Contar elementos dentro de la lista es len
```

# Bucle iterativo
Cuando queremos repetir una accion una cierta cantidad de veces, podemos utilizar un bucle iterativo 
Iteraciones

```python
dias_semana = [1, 2, 3, 4, 5, 6, 7]
for i in dias_semana:
    print(i)
    
```

```python
lista_ingredientes = ["Huevo", "Azucar", "Harina", "Polvo de Hornear", "Leche"]
for elementos in lista_ingredientes:
    print(elementos)
    
```

```python
numeros = [1*2, 2*2, 3*2, 4*2, 5*2, 6*2, 7*2, 8*2, 9*2, 10*2]
for multiplos in numeros:
    print(multiplos)
```

```python
marcas = ["Intel", "Nvidia", "Radeon", "Electrolux","Power Rangers"] 
for elementos in marcas:
    if len(elementos) >=6:
        print(elementos)
```
